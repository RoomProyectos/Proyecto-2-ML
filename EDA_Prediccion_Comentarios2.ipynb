### EDA y preprocesamiento

```python
import nltk
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
import nltk
# nltk.download()
import re
from textblob import Word
```

```python
df_completo = pd.read_csv("Data/100178_Comentarios.csv", sep=";")
df_completo.shape
```

```python
df_completo.head()
```

```python
df_completo["texto"] = df_completo["texto"].str.lower()
```

```python
df_completo.head()
```

```python
df_completo.isna().sum()
```

```python
df_no_nan = df_completo.dropna()
```

```python
df_no_nan.isna().sum()
```

```python
df_no_nan.shape
```

```python
df_no_nan.head()
```

```python
def limpiar_texto(frase):
    patron = r'[^a-zA-Z0-9\s]'  
    return ''.join(re.split(patron, frase))  


df_no_nan["texto"] = df_no_nan["texto"].apply(limpiar_texto)
```

```python
df_no_nan.head()
```

```python
# Tokenizamos el texto

df_no_nan["texto"] = df_no_nan["texto"].apply(lambda text: nltk.word_tokenize(text = text, language = "spanish"))
```

```python
df_no_nan.head()
```

```python
# Eliminamos stopwords

STOPWORDS = nltk.corpus.stopwords.words("spanish")

def is_valid(token):
    if len(token) <= 2: # Palabras muy cortas y signos de puntuación fuera
        return False
    if token in STOPWORDS: # Stopwords fuera
        return False
    return True

df_no_nan["texto"] = df_no_nan["texto"].apply(lambda tokens: [token for token in tokens if is_valid(token)])
```

```python
df_no_nan.head()
```

```python

# Lematización

lemmatizer = nltk.stem.WordNetLemmatizer()

df_no_nan["texto"] = df_no_nan["texto"].apply(lambda tokens: [lemmatizer.lemmatize(token, pos="v") for token in tokens])
```
